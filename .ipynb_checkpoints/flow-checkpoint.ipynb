\begin{equation}
    \mu\nabla^2\mathbf{u}=\nabla p-\mathbf{F}    
\end{equation}
\begin{equation}
    \nabla\cdot\mathbf{u}=0
\end{equation}

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import interpolate, sparse
from scipy.sparse import linalg
```

```python
N = 100
dx = 0.01
```

```python
_A_1 = sparse.eye(N)
D_2 = sparse.eye(N, k=1) + sparse.eye(N, k=-1) - 2 * _A_1
D_2 = D_2.tolil()
D_2[0, -1] = 1
D_2[-1, 0] = 1
L = 1 / (dx**2) * (sparse.kron(D_2, _A_1) + sparse.kron(_A_1, D_2))
```

```python
D1 = 0.5 * sparse.eye(N, k=1) - 0.5 * sparse.eye(N, k=-1)
D1 = D1.tolil()
D1[-1, 0] = 0.5
D1[0, -1] = -0.5
D1 /= dx
Dx = sparse.kron(sparse.eye(N), D1)
Dy = sparse.kron(D1, sparse.eye(N))
```

```python
yy, xx = np.mgrid[0:N, 0:N]
x = xx.flatten()
y = yy.flatten()
```

```python
def Force_x(x, y, sigma=10):
    return (
        100 * np.exp((-((x - N / 2) ** 2) - (y - N) ** 2) / sigma**2)
        + 100 * np.exp((-((x - N / 2) ** 2) - (y) ** 2) / sigma**2)
        - 100 * np.exp((-((x - N / 2) ** 2) - (y - N / 2) ** 2) / sigma**2)
    )


def Force_y(x, y, sigma=1000):
    return 0 * y  # (x-N/3)+(y-N/3)
```

```python
def flow_simulation(x, y, f_x=Force_x, f_y=Force_y, Dx=Dx, Dy=Dy, L=L):
    b_ux = -f_x(x, y)
    b_uy = -f_y(x, y)
    u_star_x = linalg.spsolve(L, b_ux)
    u_star_y = linalg.spsolve(L, b_uy)
    DXX = Dx @ u_star_x
    DYY = Dy @ u_star_y
    phi = linalg.spsolve(L, (-(DXX + DYY)))
    ux = u_star_x + Dx @ phi
    uy = u_star_y + Dy @ phi
    ux -= ux.mean()
    uy -= uy.mean()
    return ux, uy
```

```python
ux, uy = flow_simulation(x, y)
```

```python
plot = False
if plot:
    plt.figure(figsize=(18, 12))
    plt.quiver(xx, yy, ux, uy)
    # plt.streamplot(xx, yy, u_star1, u_star2)
    plt.show()
```

```python
plot = False
if plot:
    plt.figure(figsize=(18, 12))
    plt.quiver(xx, yy, Force_x(x, y), Force_y(x, y))
    # plt.streamplot(xx, yy, u_star1, u_star2)
    plt.show()
```

```python
if plot:
    plt.figure(figsize=(18, 12))
    # plt.quiver(xx, yy, ux, uy)
    plt.streamplot(xx, yy, ux.reshape(N, N), uy.reshape(N, N))
    plt.show()
```

```python
u_eff_x = interpolate.interp2d(x, y, ux)
u_eff_y = interpolate.interp2d(x, y, uy)
time_step = 0.1
lenght = int((100 - 0) / time_step)


def ball(t_step=time_step, t_stop=100, t_start=0, u_eff_x=u_eff_x, u_eff_y=u_eff_y):
    lenght = int((t_stop - t_start) / t_step)
    pos = np.empty((2, lenght + 1))
    pos[0, 0] = np.random.rand() * N
    pos[1, 0] = np.random.rand() * N
    for i in range(lenght):
        pos[0, i + 1] = pos[0, i] + u_eff_x(pos[0, i], pos[1, i]) * t_step
        if pos[0, i + 1] >= N:
            pos[0, i + 1] -= N
        elif pos[0, i + 1] <= 0:
            pos[0, i + 1] += N
        pos[1, i + 1] = pos[1, i] + u_eff_y(pos[0, i], pos[1, i]) * t_step
        if pos[1, i + 1] >= N:
            pos[1, i + 1] -= N
        elif pos[1, i + 1] <= 0:
            pos[1, i + 1] += N
    return pos
```

```python
pos = ball()
```

```python
from joblib import Parallel, delayed

Npos = 100
pos_arr = np.array(Parallel(n_jobs=-2)(delayed(ball)() for i in range(Npos)))
```

```python
def Force2_x(x, y, sigma=0.001):
    n = len(x)
    return 100 * np.random.rand(
        n
    )  # 1* np.exp((-((x - N / 2) ** 2) - ((y - N / 2) ** 2)) / sigma)


def Force2_y(x, y, sigma=1000):
    n = len(x)
    return 100 * np.random.rand(
        n
    )  # 10000 * np.exp((-((x - N / 2) ** 2) - ((y - N / 2) ** 2)) / sigma)
```

```python
ux2, uy2 = flow_simulation(x, y, f_x=Force2_x, f_y=Force2_y)
u_eff_x2 = interpolate.interp2d(x, y, ux2)
u_eff_y2 = interpolate.interp2d(x, y, uy2)
```

```python
plt.figure(figsize=(18, 12))
plt.quiver(xx, yy, ux2, uy2)
#plt.streamplot(xx, yy, ux.reshape(N, N), uy.reshape(N, N))
#plt.plot(pos[0], pos[1], ".")
plt.show()
```

```python
plt.figure(figsize=(18, 12))
plt.quiver(xx, yy, Force2_x(x,y), Force2_y(x,y))
#plt.streamplot(xx, yy, ux.reshape(N, N), uy.reshape(N, N))
#plt.plot(pos[0], pos[1], ".")
plt.show()
```

```python
Npos = 100
pos_arr_2 = np.array(
    Parallel(n_jobs=-2)(
        delayed(ball)(u_eff_x=u_eff_x2, u_eff_y=u_eff_y2) for i in range(Npos)
    )
)
```

```python
%%time
import matplotlib.animation as animation

Npos_ani = 10

fps = 100
nSeconds = time_step

fig = plt.figure(figsize=(8, 8))
plt.streamplot(xx, yy, ux.reshape(N, N), uy.reshape(N, N))
(ln,) = plt.plot([], [], "r.")


def animate_func(i):
    plt.title(f"t = {(i*nSeconds):.2f}")
    ln.set_data(
        [pos_arr[j, 0, i] for j in range(Npos_ani)],
        [pos_arr[k, 1, i] for k in range(Npos_ani)],
    )
    return ln


anim = animation.FuncAnimation(
    fig,
    animate_func,
    frames=lenght,
    interval=1 / fps,
)

anim.save("anim23_test.mp4", fps=fps, extra_args=["-vcodec", "libx264"])

print("Done!")
```
