```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import sparse
from scipy.sparse import linalg
```

```python
N = 100
dx = 0.001
```

```python
left_x = 1
left_y = -2
top_x = -2
top_y = 1
```

```python
bottom_x = top_x
bottom_y = top_y
right_x = left_x
right_y = left_y
```

```python
_A_1 = sparse.eye(N)
D_2 = sparse.eye(N, k=1) + sparse.eye(N, k=-1) - 2 * _A_1
L = 1 / (dx**2) * (sparse.kron(D_2, _A_1) + sparse.kron(_A_1, D_2))
b_ux = np.zeros(N**2)
b_uy = b_ux.copy()
```

```python
yy, xx = np.mgrid[0:N, 0:N]
x = xx.flatten()
y = yy.flatten()
```

```python
def Force_x(x, y, sigma=0.001):
    n = len(x)
    return np.random.rand(n) #1* np.exp((-((x - N / 2) ** 2) - ((y - N / 2) ** 2)) / sigma)


def Force_y(x, y, sigma=1000):
    n = len(x)
    return np.random.rand(n) #10000 * np.exp((-((x - N / 2) ** 2) - ((y - N / 2) ** 2)) / sigma)
```

```python
b_ux = -Force_x(x, y)/dx
b_uy = -Force_y(x, y)/dx
```

```python
L1 = L.copy()
L1 = L1.tolil()
for i in np.arange(0, N**2, N):
    L1[i] = 0
    L1[i, i] = 1
    b_ux[i] = left_x
    b_uy[i] = left_y

for i in np.arange(N - 1, N**2, N):
    L1[i] = 0
    L1[i, i] = 1
    b_ux[i] = right_x
    b_uy[i] = right_y

for i in range(N):
    L1[i] = 0
    L1[i, i] = 1
    b_ux[i] = bottom_x
    b_uy[i] = bottom_y
    _i = -i - 1
    L1[_i] = 0
    L1[_i, _i] = 1
    b_ux[_i] = top_x
    b_uy[_i] = top_y
```

```python
L1 = L1.tocsr()
```

```python
u_star_x = linalg.spsolve(L1, b_ux)
u_star_y = linalg.spsolve(L1, b_uy)
```

```python
plt.figure(figsize=(18, 12))
plt.quiver(xx, yy, u_star_x, u_star_y)
plt.show()
```

```python
D1 = 0.5 * sparse.eye(N, k=1) - 0.5 * sparse.eye(N, k=-1)
D1 = D1.tolil()
D1[0, 0] = -1
D1[0, 1] = 1
D1[-1, -1] = 1
D1[-1, -2] = -1
D1 /= dx
Dx = sparse.kron(sparse.eye(N), D1)
Dy = sparse.kron(D1, sparse.eye(N))
```

```python
DXX = Dx @ u_star_x
DYY = Dy @ u_star_y
```

```python
phi = linalg.spsolve(L, (-(DXX + DYY)))
```

```python
dphi = [Dx @ phi, Dy @ phi]
```

```python
ux = u_star_x + dphi[0]
uy = u_star_y + dphi[1]
```

```python
plt.figure(figsize=(18, 12))
plt.quiver(xx, yy, ux, uy)
# plt.streamplot(xx, yy, u_star1, u_star2)
plt.show()
```

```python
plt.figure(figsize=(18, 12))
#plt.quiver(xx, yy, ux, uy)
plt.streamplot(xx, yy, ux.reshape(N,N), uy.reshape(N,N))
plt.show()
```
